````python
import os
import json
import time
import pandas as pd
import google.generativeai as genai
from dotenv import load_dotenv

# --- 1. โหลด API Key จาก .env ---
load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")

if not API_KEY:
    raise ValueError("ไม่พบ GEMINI_API_KEY ในไฟล์ .env กรุณาตรวจสอบอีกครั้ง")

genai.configure(api_key=API_KEY)

# กำหนดโครงสร้าง Model และ System Instruction
model = genai.GenerativeModel(
    model_name="gemini-3-flash-preview",
    system_instruction="""คุณคือผู้เชี่ยวชาญด้านวิจัยตลาด (Market Researcher) 
    หน้าที่ของคุณคือวิเคราะห์ความคิดเห็นลูกค้าและจัดกลุ่ม (Tagging) ตามหัวข้อที่กำหนด 
    โดยตอบกลับเป็น JSON Object เท่านั้น (1=เกี่ยวข้อง, 0=ไม่เกี่ยวข้อง) 
    ห้ามตอบเป็นประโยคอธิบายอื่นๆ"""
)

target_topics = [
        "Sensory_Quality", "Psychological_Acceptance", "Packaging_Visual",
        "Health_Nutrition", "Safety_Trust", "Price_Value", "Convenience_Format", "ไม่มีความเกี่ยวข้องกับผลิตภัณฑ์"
    ]


# --- 2. ฟังก์ชันวิเคราะห์ (รองรับ Multi-label) ---
def analyze_comment(text, topics):
    prompt = f"""
    วิเคราะห์ข้อความต่อไปนี้โดยใช้หัวข้อ: {topics}
    
    ข้อความ: "{text}"
    
    รูปแบบการตอบ (JSON):
        {{
            "{topics[0]}": 0 หรือ 1,
            "{topics[1]}": 0 หรือ 1,
            ...
        }}
    """
    try:
        response = model.generate_content(prompt)
        # ลบ Markdown backticks ออกถ้ามี
        clean_json = response.text.replace('```json', '').replace('```', '').strip()
        return json.loads(clean_json)
    except Exception as e:
        print(f"เกิดข้อผิดพลาดในการวิเคราะห์: {e}")
        return {topic: 0 for topic in topics}

# --- 3. ขั้นตอนการรัน (Main Process) ---

# โหลด Data
# input_file = 'comments.csv' # เปลี่ยนชื่อไฟล์ตามจริงของคุณ
# df = pd.read_csv(input_file)


all_labels = []
        
# print(f"เริ่มประมวลผลข้อมูลจำนวน {len(df)} รายการ...")

# for index, row in df.iterrows():
#             comment = row['comment']
#             print(f"[{index+1}/{len(df)}] วิเคราะห์ ID: {row['customer_Id']}")
            
#             # ส่งไปให้ Gemini วิเคราะห์
#             result = analyze_comment(comment, target_topics)
#             all_labels.append(result)
            
#             # หน่วงเวลาสำหรับ Free Tier (ป้องกัน Rate Limit)
#             time.sleep(2)

# # รวมผลลัพธ์เข้ากับ DataFrame เดิม
# label_df = pd.DataFrame(all_labels)
# final_df = pd.concat([df, label_df], axis=1)

# # บันทึกไฟล์
# output_file = 'final_labeled_results.csv'
# final_df.to_csv(output_file, index=False, encoding='utf-8-sig')
# print(f"\nเสร็จสิ้น! บันทึกผลไปที่: {output_file}")

result = analyze_comment("คิดว่าเรื่องแพ็กเกจเป็นเรื่องสำคัญอันดับหนึ่งเพราะเป็นสิ่งแรกที่คนเห็น", target_topics)
all_labels.append(result)

````
