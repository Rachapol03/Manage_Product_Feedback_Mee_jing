````python
import pandas as pd
````

````python
df = pd.read_csv('model_จิ้งหรีดV1 - merged_data.csv')
df['customer_Id'] = df.index+1
df['customer_Id']
df.head()
````

````python
dffeeddata = df.copy()
dffeeddata = dffeeddata[['customer_Id', 'ถ้าคุณมีโอกาสแนะนำ คุณอยากบอกอะไรเพื่อให้ทำผลิตภัณฑ์นี้ออกมาได้ถูกใจคุณมากที่สุด?']]
dffeeddata.dropna(inplace=True)
dffeeddata = dffeeddata.reset_index(drop=True)
dffeeddata.head()
````

````python
import os
import json
import time
import pandas as pd
import google.generativeai as genai
from dotenv import load_dotenv

# --- 1. โหลด API Key จาก .env ---
load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")

if not API_KEY:
    raise ValueError("ไม่พบ GEMINI_API_KEY ในไฟล์ .env กรุณาตรวจสอบอีกครั้ง")

genai.configure(api_key=API_KEY)

# กำหนดโครงสร้าง Model และ System Instruction
model = genai.GenerativeModel(
    model_name="gemini-2.0-flash",
    system_instruction="""คุณคือผู้เชี่ยวชาญด้านวิจัยตลาด (Market Researcher) 
    หน้าที่ของคุณคือวิเคราะห์ความคิดเห็นลูกค้าและจัดกลุ่ม (Tagging) ตามหัวข้อที่กำหนด 
    โดยตอบกลับเป็น JSON Object เท่านั้น (1=เกี่ยวข้อง, 0=ไม่เกี่ยวข้อง) 
    ห้ามตอบเป็นประโยคอธิบายอื่นๆ"""
)

target_topics = [
        "Sensory_Quality", "Psychological_Acceptance", "Packaging_Visual",
        "Health_Nutrition", "Safety_Trust", "Price_Value", "Convenience_Format", "Unrelated to the product"
    ]


# --- 2. ฟังก์ชันวิเคราะห์ (รองรับ Multi-label) ---
def analyze_comment(text, topics):
    prompt = f"""
    วิเคราะห์ข้อความต่อไปนี้โดยใช้หัวข้อ: {topics}
    
    ข้อความ: "{text}"
    
    รูปแบบการตอบ (JSON):
        {{
            "{topics[0]}": 0 หรือ 1,
            "{topics[1]}": 0 หรือ 1,
            ...
        }}
    """
    try:
        response = model.generate_content(prompt)
        clean_json = response.text.replace('```json', '').replace('```', '').strip()
        return json.loads(clean_json)
    except Exception as e:
        print(f"เกิดข้อผิดพลาดในการวิเคราะห์: {e}")
        return {topic: 0 for topic in topics}



all_labels = []
print(f"เริ่มประมวลผลข้อมูลจำนวน {len(dffeeddata)} รายการ...")

for index, row in dffeeddata.iterrows():
            comment = row['ถ้าคุณมีโอกาสแนะนำ คุณอยากบอกอะไรเพื่อให้ทำผลิตภัณฑ์นี้ออกมาได้ถูกใจคุณมากที่สุด?']
            print(f"[{index+1}/{len(dffeeddata)}] วิเคราะห์ ID: {row['customer_Id']}")
            
            # ส่งไปให้ Gemini วิเคราะห์
            result = analyze_comment(comment, target_topics)
            all_labels.append(result)
            
            # หน่วงเวลาสำหรับ Free Tier (ป้องกัน Rate Limit)
            time.sleep(5)

# รวมผลลัพธ์เข้ากับ DataFrame เดิม
label_df = pd.DataFrame(all_labels)
final_df = pd.concat([dffeeddata, label_df], axis=1)

# # บันทึกไฟล์
# output_file = 'final_labeled_results.csv'
# final_df.to_csv(output_file, index=False, encoding='utf-8-sig')
# print(f"\nเสร็จสิ้น! บันทึกผลไปที่: {output_file}")

# result = analyze_comment("คิดว่าเรื่องแพ็กเกจเป็นเรื่องสำคัญอันดับหนึ่งเพราะเป็นสิ่งแรกที่คนเห็น", target_topics)
# all_labels.append(result)
final_df.head()

# --- รวม final_df เข้ากับ df หลัก โดยใช้ customer_Id ---
# สร้าง merged_df เพื่อไม่ให้ทับข้อมูลต้นฉบับ ถ้าต้องการทับให้ใช้ df = merged_df
merged_df = df.merge(final_df, on='customer_Id', how='left')
print(f"Before merge df shape: {df.shape}")
print(f"final_df shape: {final_df.shape}")
print(f"merged_df shape: {merged_df.shape}")
merged_df.head()
````
